# Pandas Date and Time

Load data.

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

So far all we've learned when handling dates is to extract the date. But what if we wanted to extract other information? Like month or year?

## Extract Month

### Notes

* `dt.month`: extract the month from a datetime in a DataFrame.
* `dt`: accessor that provides a way to access specialized methods and properties we use to work with datatime data within a pandas series. 
* `month`: extract the month component from the datetime object in the series. 
* Use together `dt.month` on our series.

### Example

We are going to extract the month from the `job_posted_date` column.

```python
df['job_posted_month'] = df['job_posted_date'].dt.month
```

```python
df['job_posted_month']
```

Now if we compare it with the `job_posted_date` column using `iloc[]` to get the first 10 rows and first 8 columns of our DataFrame we can see that it's correctly getting the month from the column.

*Reminder*: `job_posted_date` is in this format: `YYYY-MM-DD HH:MM:SS`.

```python
df.iloc[:10,:9]
```

```python
df['job_posted_year'] = df['job_posted_date'].dt.year

df['job_posted_year']
```
