```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
from ast import literal_eval

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda x:literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', startangle=90, autopct ='%1.1f%%')
plt.title('Work From Home')
plt.ylabel('')
plt.show()
```

```python
df[['job_work_from_home','job_no_degree_mention','job_health_insurance']]
```

```python
fig,ax=plt.subplots(1,3)
dict_columns = {'job_work_from_home':'Work from Home',
                'job_no_degree_mention': 'Job degree Mentioned',
                'job_health_insurance':'Job Health Insurance'}
for i,(column, title) in enumerate(dict_columns.items()):
    df[column].value_counts().plot(kind='pie',ax=ax[i],
        title=title, startangle=90, autopct='%1.1f%%')
    ax[i].set_ylabel('')

fig.tight_layout()
```
