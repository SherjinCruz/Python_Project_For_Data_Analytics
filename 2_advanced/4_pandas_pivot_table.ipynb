```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## Count oF Job Titles

```python
df.pivot_table(index='job_title_short',aggfunc='size')
```

```python
df.groupby('job_title_short')['job_title_short'].size()
```

## Median Salary Avg by Jobs

```python
df.groupby('job_title_short')['salary_year_avg'].median()
```

```python
df.pivot_table(values='salary_year_avg',index='job_title_short',aggfunc='median')
```

## Median Salary Analysis By Country

```python
df.pivot_table(index='job_country',values='salary_year_avg',columns='job_title_short',aggfunc='median')
```

## Bar Chart for Top 6 Countries

```python
top_countries = df['job_country'].value_counts().head(6).index
df_job_country_salary = df.pivot_table(
    index='job_country',
    values='salary_year_avg',
    columns='job_title_short',
    aggfunc='median'
    )
df_job_country_salary = df_job_country_salary.reindex(top_countries)
job_titles = ['Data Scientist','Data Analyst','Data Engineer']
df_job_country_salary = df_job_country_salary.reindex(columns=job_titles)
```

```python
df_job_country_salary.plot(kind='bar')
plt.title('Median Salary by Country and job Titles')
plt.ylabel('Median Salary($USD')
plt.xlabel('')
plt.xticks(rotation=45,ha='right')
plt.show()
```
