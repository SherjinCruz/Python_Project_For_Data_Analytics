## Exercise - Analyzing Data Science Job Demand

```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country'] == 'United States'].copy()
df_US
```

```python
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
```

```python
df_US
```

```python
df_US_pivot = df_US.pivot_table(index='job_posted_month',aggfunc='size',columns='job_title_short')
```

```python
df_US_pivot
```

```python
df_US_pivot.reset_index(inplace=True)
```

```python
df_US_pivot['month_number']=pd.to_datetime(df_US_pivot['job_posted_month'],format='%B').dt.month
df_US_pivot
```

```python
df_US_pivot.sort_values('month_number',inplace=True)
df_US_pivot
```

```python
df_US_pivot.set_index('job_posted_month',inplace=True)
```

```python
df_US_pivot.drop(columns='month_number',inplace=True)
```

```python
df_US_pivot
```

```python
df_US_pivot.plot(kind='line')
```

```python
top_3 = df_US['job_title_short'].value_counts().head(3)
top_3 = top_3.index.to_list()
top_3
```

```python
df_US_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Posting for Top Data Jobs')
plt.xlabel('Month')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```
