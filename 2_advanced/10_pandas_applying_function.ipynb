## apply() - Calaculate Projected Salary Next Year

```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
def projected_salary(salary):
    return salary * 1.03

df_salary = df.loc[df['salary_year_avg'].notna(),['salary_year_avg']].copy()

df_salary['salary_year_inflated'] = df_salary.apply(projected_salary)
```

```python
def projected_salary(salary):
    if 'senior' in df['job_title_short']:
        return salary * 1.5
    else:
        return salary * 1.3

df_salary = df[['job_title_short','salary_year_avg']].copy()

df_salary['salary_year_inflated'] = df_salary['salary_year_avg'].apply(projected_salary)
df_salary[~df_salary['job_title_short'].str.contains('Senior', na=False)]
```

```python
def projected_salary(row):
    if 'senior' in row['job_title_short']:
        return row['salary_year_avg'] * 1.5
    else:
        return  row['salary_year_avg'] * 1.3

df_salary = df.loc[df['salary_year_avg'].notna(),['job_title_short','salary_year_avg']].copy()

df_salary['salary_year_inflated'] = df_salary.apply(projected_salary,axis=1)
df_salary[df_salary['job_title_short'].str.contains('Senior', na=False)]
```
