```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
from ast import literal_eval

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda x:literal_eval(x) if pd.notna(x) else x)
```

```python
df
```

```python
df_skill = df.explode('job_skills')
df_skill
```

```python
skill_exploded = df_skill.groupby(['job_title_short','job_skills'])['job_skills'].size()
skill_exploded = skill_exploded.reset_index(name='skill_list')
df_title = skill_exploded.sort_values(by='skill_list',ascending=False)
df_title
```

```python
title = 'Data Scientist'
count=15
df_jobs = df_title[df_title['job_title_short']==title].head(count)
df_jobs
```

```python
df_jobs.plot(kind='barh',x='job_skills',y='skill_list')
plt.gca().invert_yaxis()
plt.title(f'Top {count} skills for {title}')
plt.xlabel('job_count')
plt.ylabel('')
plt.legend().set_visible(False)
plt.show()
```
