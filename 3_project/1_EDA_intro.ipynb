## Exploratory Data Analysis


```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset
from ast import literal_eval

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda x:literal_eval(x) if pd.notna(x) else x)
```

## Filter for US Data Analyst Roles

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')]
df_DA_US
```

```python
df_plot = df_DA_US['job_location'].value_counts().head(10).to_frame() #to_frame convert as data frame
df_plot
```

```python
sns.barplot(data=df_plot, x='count',y='job_location',hue='count',palette = 'dark:r_r',legend=False)
sns.set_theme(style='ticks')
sns.despine()
plt.title('Count of Job of Data Analyst in US')
plt.xlabel('No of Jobs')
plt.ylabel('')
plt.show()
```

```python
dict_columns = {'job_work_from_home':'Work From Home',
                'job_no_degree_mention':'Job Degree Mentioned',
                'job_health_insurance':'Job Health Insurance'}

fig,ax = plt.subplots(1,3)
fig.set_size_inches((10,5)) #increase size

for i,(column,title) in enumerate(dict_columns.items()):
    df_plot_pie = df_DA_US[column].value_counts()
    df_plot_pie.plot(kind='pie',title=title,ax=ax[i],autopct='%1.1f%%')
    ax[i].set_ylabel('')
fig.tight_layout()

```

```python
df_plot = df_DA_US['company_name'].value_counts().head(10).to_frame() #to_frame convert as data frame
sns.barplot(data=df_plot, x='count',y='company_name',hue='count',palette = 'dark:g_r',legend=False)
sns.set_theme(style='ticks')
sns.despine()   #remove square in chart
plt.title('Count of Company of Data Analyst Jobs in US')
plt.xlabel('No of Jobs')
plt.ylabel('')
plt.show()
```
