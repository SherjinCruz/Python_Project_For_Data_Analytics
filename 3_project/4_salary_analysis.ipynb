## How Well Do The Job Pay and Skills For Data Analysis

```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset
from ast import literal_eval

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda x:literal_eval(x) if pd.notna(x) else x)
```

```python
df
```

```python
df_US = df[df['job_country'] == 'United States'].copy()
df_US
```

```python
job_titles = df_US['job_title_short'].value_counts().index[:6].to_list()
job_titles
```

```python
df_US_top_6 = df_US[df_US['job_title_short'].isin (job_titles)]
df_US_top_6
```

```python
median = df_US_top_6.groupby('job_title_short')['salary_year_avg'].median()
median = median.sort_values(ascending=False).index
```

```python
sns.boxplot(data=df_US_top_6,x='salary_year_avg',y='job_title_short',order=median)
sns.set_theme(style='ticks')
#df_US_top_6.plot(kind='box',column='salary_year_avg',by='job_title_short',vert=False)
plt.ylabel('')
plt.xlabel('Yearly Salary ($USD)')
plt.xlim(0,600000)
plt.title("Salary Distribution of Data Jobs in US")
ax=plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x,pos: f'${int(x/1000)}k'))
plt.show()
```

## Investigate Median Salary with Skills on Data Analyst

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
df_DA_US = df_DA_US.explode('job_skills')
df_DA_US
```

```python
df_DA_top_pay = df_DA_US.groupby('job_skills')['salary_year_avg'].agg(['count','median'])
df_DA_top_pay = df_DA_top_pay.sort_values(by='median',ascending=False).head(10)
df_DA_top_pay 
```

```python
df_DA_skills= df_DA_US.groupby('job_skills')['salary_year_avg'].agg(['count','median'])
df_DA_skills = df_DA_skills.sort_values(by='count',ascending=False).head(10)
df_DA_skills
```

```python
fig,ax=plt.subplots(2,1)
sns.set_theme(style='ticks')
sns.barplot(data=df_DA_top_pay,x='median',y=df_DA_top_pay.index,hue='median',palette='dark:g_r',ax=ax[0],legend=False)
ax[0].set_title('Top 10 Highest Paid Skills for Data Analyst')
ax[0].set_xlabel('Yearly Salary ($USD)')
ax[0].set_ylabel('')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter (lambda x,pos: f'${int(x/1000)}K'))

sns.barplot(data=df_DA_skills,x='count',y=df_DA_top_pay.index,hue='count',palette='dark:r_r',ax=ax[1])
ax[1].set_title('Top 10 Most In Demand Skills for Data Analyst')
ax[1].set_xlabel('Count')
ax[1].set_ylabel('')
ax[1].legend().remove()
sns.despine()
plt.tight_layout()
```
